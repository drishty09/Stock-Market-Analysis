### **1. Fetching Stock Market Data using yfinance**

```python
import yfinance as yf

# Fetch stock data
ticker = "AAPL"
stock_data = yf.download(ticker, period="1y", interval="1d")

# Display basic info
print(stock_data.head())
print(stock_data.info())
```

### **2. Visualizing Stock Price Data**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(stock_data["Close"], label="AAPL Adjusted Close Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Apple Stock Price Over Time")
plt.grid()
plt.legend()
plt.show()
```

### **3. Fetching Fundamental Data**

```python
msft = yf.Ticker("MSFT")

# Get financials
revenue = msft.financials.loc["Total Revenue"]
ebit = msft.financials.loc["EBIT"]

# Plot revenue and EBIT
revenue.plot(kind="bar", title="Microsoft Revenue")
plt.show()

ebit.plot(kind="bar", title="Microsoft EBIT")
plt.show()
```

### **4. Resampling Stock Data**

```python
stock_data_resampled = stock_data.resample("H").mean()
print(stock_data_resampled.head())
```

### **Statistical Analysis Of Stock Data**

```python
import yfinance as yf

# Replace 'AAPL' with your stock ticker
ticker = "AAPL"
stock_data = yf.download(ticker, period="1y", interval="1d")

# Descriptive statistics
mean_price = stock_data["Close"].mean()
median_price = stock_data["Close"].median()
std_dev = stock_data["Close"].std()

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Standard Deviation: {std_dev}")
```

```python
stock = yf.Ticker(ticker)

pe_ratio = stock.info.get("trailingPE", "N/A")
pb_ratio = stock.info.get("priceToBook", "N/A")

print(f"P/E Ratio: {pe_ratio}")
print(f"P/B Ratio: {pb_ratio}")
```

```python
revenue = stock.financials.loc["Total Revenue"].iloc[0]
ebit = stock.financials.loc["EBIT"].iloc[0]

print(f"Total Revenue: {revenue}")
print(f"EBIT: {ebit}")
```
